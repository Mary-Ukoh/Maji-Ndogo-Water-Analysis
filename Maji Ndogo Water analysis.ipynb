```python
%%sql
SELECT 
     town_name,
    COUNT(town_name) AS record_per_town
FROM location
GROUP BY town_name
ORDER BY record_per_town DESC;
```

```python
%%sql
SELECT 
     province_name,
    COUNT(province_name) AS record_per_town
FROM location
GROUP BY province_name 
ORDER BY record_per_town DESC;
```

```python
/* find a way to do the following:
1. Create a result set showing:
• province_name
• town_name
• An aggregated count of records for each town (consider naming this records_per_town).
• Ensure the data is grouped by both province_name and town_name.
2. Order the results primarily by province_name. Within each province, 
further sort the towns by their record counts in descending order.

*/
```

```python
%%sql
SELECT
     town_name,
     province_name,
    COUNT(town_name) AS record_per_town,
    COUNT(province_name) AS record_per_province
FROM location
GROUP BY province_name,town_name 
ORDER BY province_name, record_per_town DESC, record_per_province DESC;
```

```python

%%sql
SELECT 
SUM(number_of_people_served) AS Total_people_served

FROM water_source;
```

```python

%%sql
SELECT 
type_of_water_source,
COUNT(type_of_water_source) AS count_of_water_source

FROM water_source
GROUP BY type_of_water_source
ORDER BY count_of_water_source DESC;

```

```python
%%sql
SELECT 
type_of_water_source,
ROUND(AVG(number_of_people_served)) AS Avg_number_of_people_served

FROM water_source
GROUP BY type_of_water_source
ORDER BY Avg_number_of_people_served DESC;
```

```python
%%sql
SELECT 
type_of_water_source,
ROUND(SUM(number_of_people_served)) AS Sum_number_of_people_served

FROM water_source
GROUP BY type_of_water_source
ORDER BY Sum_number_of_people_served DESC;
```

```python
%%sql
SELECT type_of_water_source, 
ROUND(SUM(number_of_people_served)/
(SELECT SUM(number_of_people_served) FROM water_source)* 100) AS percentage_of_people_served 
FROM water_source
GROUP BY type_of_water_source
ORDER BY percentage_of_people_served DESC;
```

```python
%%sql
SELECT 
SUM(number_of_people_served)* 100
FROM water_source;
```

```python
%%sql
SELECT 
type_of_water_source,
SUM(number_of_people_served) AS Sum_num_of_people_served,
RANK() OVER (ORDER BY Sum(number_of_people_served)) AS Ranking
FROM water_source
WHERE type_of_water_source != 'tap_in_home'
GROUP BY type_of_water_source;
```

```python
%%sql

SELECT source_id, 
type_of_water_source, 
(number_of_people_served) AS num_of_people_served, 
RANK() OVER (PARTITION BY type_of_water_source 
         ORDER BY number_of_people_served DESC) AS Priority_of_ranking
FROM water_source
WHERE type_of_water_source != "tap_in_home"
LIMIT 1000;
```

```python
%%sql
SELECT source_id, type_of_water_source,  number_of_people_served AS num_of_people_served, 
DENSE_RANK() OVER (PARTITION BY type_of_water_source 
         ORDER BY number_of_people_served ) AS Priority_of_ranking
FROM water_source
WHERE type_of_water_source != "tap_in_home"
LIMIT 1000;
```

```python
%%sql
SELECT DAYNAME(time_of_record) AS Day_of_the_week, 
AVG(time_in_queue) AS Avg_queue_time
FROM visits
GROUP BY DAYNAME(time_of_record);

```

```python
%%sql
SELECT HOUR(time_of_record) AS Hour_per_day,
ROUND(AVG(time_in_queue)) AS Avg_queue_time
FROM visits
GROUP BY HOUR(time_of_record)
ORDER BY Hour_per_day DESC;
```

```python
%%sql
SELECT
TIME_FORMAT(TIME(time_of_record), '%H:00') AS hour_of_day,
ROUND(AVG(time_in_queue)) AS Avg_queue_time
FROM visits
GROUP BY TIME_FORMAT(TIME(time_of_record), '%H:00')
;
```

```python
%%sql
SELECT
TIME_FORMAT(TIME(time_of_record), '%H:00') AS hour_of_day,
DAYNAME(time_of_record),
CASE
WHEN DAYNAME(time_of_record) = 'Sunday' THEN time_in_queue
ELSE NULL
END AS Sunday
FROM
visits
WHERE
time_in_queue != 0
LIMIT 100;
```

```python
%%sql
SELECT
TIME_FORMAT(TIME(time_of_record), '%H:00') AS hour_of_day,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Sunday' THEN time_in_queue
ELSE NULL
END
),0) AS Sunday,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Monday' THEN time_in_queue
ELSE NULL
END
),0) AS Monday,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Tuesday' THEN time_in_queue
ELSE NULL
END
),0) AS Tuesday,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Wednesday' THEN time_in_queue
ELSE NULL
END
),0) AS Wednesday,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Thursday' THEN time_in_queue
ELSE NULL
END
),0) AS Thursday,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Friday' THEN time_in_queue
ELSE NULL
END
),0) AS Friday,

ROUND(AVG(
CASE
WHEN DAYNAME(time_of_record) = 'Saturday' THEN time_in_queue
ELSE NULL
END
),0) AS Saturday

FROM
visits
WHERE
time_in_queue != 0 
GROUP BY
hour_of_day
ORDER BY
hour_of_day;
```
